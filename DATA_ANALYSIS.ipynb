```python
df['Launch Mass (kg.)'] = df['Launch Mass (kg.)'].replace(" ", 0).astype('float64')
df['Launch Mass (kg.)']
```

```python
a = df[
    ["Launch Site", "Launch Vehicle","Name of Satellite, Alternate Names","Purpose"]
]
a
```

```python
colors = []
for index, row in df.iterrows():
    genre = row["Launch Vehicle"]
    if genre == "PSLV":
        colors.append("blue")
    elif genre == "GSLV":
        colors.append("purple")
    elif genre == "SAriane 5":
        colors.append("orange")
    else:
        colors.append("gray")
```

```python
fig = plt.figure(figsize=(12, 8))
plt.scatter(
    df["Apogee (km)"], df["Launch Mass (kg.)"], c=colors, marker="o"
)
plt.title("Apogee and Launch Mass of a Rocket")
plt.xlabel("Apogee in KM")
plt.ylabel("Launching Mass in KG")
plt.show()
```
